### Exploring sentinel data

```python
import os
import shutil
import zipfile

#with zipfile.ZipFile('ProjData.zip', 'r') as zip_ref:
#    zip_ref.extractall()
```

```python
import numpy as np
import rasterio
from rasterio.plot import show
import matplotlib.pyplot as plt
%matplotlib inline
```

### prep functions

```python
def loadband(path):
    with rasterio.open(path) as src:
           return src.read()
        
def calculate_ndvi(nirband,redband):
    nirband = nirband.astype(np.float64)
    redband = redband.astype(np.float64)
    ndvi = (nirband - redband)/(nirband + redband+0.0000000001)
    return ndvi

def calculate_ndwi(greenband,nirband):
    greenband = greenband.astype(np.float64)
    nirband = nirband.astype(np.float64)
    ndwi = (greenband - nirband)/(greenband + nirband+0.0000000001)
    return ndwi

def save_img(filename,img_data,metadata_src,dest_folder):
    with rasterio.open(metadata_src) as src:
        kwargs = src.meta.copy()
        kwargs.update({ 'dtype': rasterio.float64 })
    if not os.path.exists(dest_folder):
        os.makedirs(dest_folder)
    with rasterio.open(dest_folder+'/'+filename+'.tif','w', **kwargs) as dst:
        dst.write(img_data)
        
def quick_viz(band):
    bandnew = np.ma.masked_where(band == 0 ,
                          band,
                          copy=True)
    fig = plt.figure(figsize= (6,6))
    rasterio.plot.show(bandnew)
    
def band_stats(band): 
    stats = []
    stats.append({
            'min':band.min(),
            'mean':band.mean(),
            'median':np.median(band),
            'max':band.max()})
    print(stats)    
```

```python
import imageio
#import imageio.v2 as imageio
import geopandas as gpd
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.pyplot as mplot
import matplotlib.image as mping
import matplotlib.colors as colors
from matplotlib.colors import ListedColormap
```

```python
cwd = os.getcwd()
```

```python
input_dir1 = 'ndvi_imgs'
os.chdir(input_dir1)
output_dir = 'animated/'
if not os.path.exists(output_dir): os.makedirs(output_dir) 
```

```python

ndvi_list = [file for file in os.listdir() if file.endswith('.tif')]

for img in ndvi_list:
    with rasterio.open(img) as src:
        lcpri_img = src.read(1)
    lcpri_img = np.ma.masked_where(lcpri_img == 0,lcpri_img,copy=True)
    cmap = 'YlGn'
    cmap_reversed = matplotlib.cm.get_cmap('YlGn_r')
    plt.figure(figsize=(10,10))
    plt.imshow(lcpri_img, cmap=cmap,vmin = -1.0, vmax=1.0)
    parts = img.split('.')
    filename = parts[0]
    print('processing: {}'.format(filename))
    plt.title(filename, fontsize = 20, fontweight = 'bold')
    cur_axes = mplot.gca()
    cur_axes.axes.get_xaxis().set_visible(False)
    cur_axes.axes.get_yaxis().set_visible(False)
    cb = plt.colorbar()
    cb.set_label('NDVI', size = 20)
    cb.ax.tick_params(labelsize=18)
    plt.tight_layout()
    plt.savefig('{}{}.png'.format(output_dir,filename, dpi=150))
    plt.close()
```

```python
png_dir = 'animated'
images = []
for file_name in os.listdir(png_dir):
    if file_name.endswith('.png'):
        file_path = os.path.join(png_dir, file_name)
        images.append(imageio.imread(file_path))

# Make it pause at the end so that the viewers can ponder
#for _ in range(10):
#    images.append(imageio.imread(file_path))


imageio.mimsave('animated/ndvi.gif', images, duration = 0.90)
```
